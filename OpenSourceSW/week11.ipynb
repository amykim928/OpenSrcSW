```python
#실습 9번
class Stack:
    def _init_(self):
        self.data = []
    def size(self):
        return len(self.data)
    def isEmpty(self):
        return self.size() == 0
    def push(self,item):
        self.data.append(item)
    def pop(self):
        return self.data.pop()
    def peek(self):
        return self.data[-1]
    def splitNumOp(self,s):
        list = []
        num = 0
        numCheck = False

        for i in s:
            if i==' ':
                continue
            if i in '0123456789':
                num = num*10 + int(i)
                numCheck = True
            else:
                if numCheck:
                    list.append(num)
                    num = 0
                numCheck = False
                list.append(i)
        
        if numCheck:
            list.append(num)
        return list
    
    def changePostFix(self,list):
        dic = {'*':5,'/':5,'+':3,'-':3,'(':1}
        opStack = Stack()
        opStack._init_()
        newList = []
        
        for i in list:
            if type(i) is int:
                newList.append(i)
            elif i == '(':
                opStack.push(i)
            elif i == ')':
                while newStack.peek() != '(':
                    newList.append(opStack.pop())
                opStack.pop()
            else:
                if opStack.isEmpty():
                    opStack.push(i)
                else:
                    while opStack.size() > 0:
                        if dic[opStack.peek()]>=dic[i]:
                            newList.append(opStack.pop())
                        else:
                            break
                    opStack.push(i)
        
        while opStack.isEmpty() == False:
            newList.append(opStack.pop())
            
        return newList
    
    def postFixCal(self,list):
        calStack = Stack()
        calStack._init_()
        
        for i in list:
            if type(i) is int:
                calStack.push(i)
            elif i == '+':
                num1 = calStack.pop()
                num2 = calStack.pop()
                calStack.push(num2+num1)
            elif i == '-':
                num1 = calStack.pop()
                num2 = calStack.pop()
                calStack.push(num2-num1)
            elif i == '*':
                num1 = calStack.pop()
                num2 = calStack.pop()
                calStack.push(num2*num1)
            elif i == '/':
                num1 = calStack.pop()
                num2 = calStack.pop()
                calStack.push(int(num2/num1))
        return calStack.pop()
        
    def printCal(self,list):
        
        newStr = ''
        for i in list:
            newStr += "{}".format(i)
        
        return newStr
    
stack = Stack()
str = input("계산식 입력: ")
postList = stack.splitNumOp(str)
calList = stack.changePostFix(postList)
result = stack.postFixCal(calList)
printList = stack.printCal(postList) + "=" + "{}".format(result)
print("계산식 출력: " + printList)
```
