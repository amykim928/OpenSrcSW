```python
#실습 1번
import csv
import matplotlib.pyplot as plt

f = open('seoulTemp.csv') 
#seoulTemp.csv은 1904년 1월 1일부터 2020년 12월 31일까지 서울 지역의 기온 데이터를 포함한 파일입니다.
data = csv.reader(f)
next(data)
high = []
low = []

for row in data:
    if row[-1]!='' and row[-2]!='':
        if row[0].split('-')[1]=='09' and row[0].split('-')[2]=='28':
            high.append(float(row[-1]))
            low.append(float(row[-2]))
            
plt.plot(high, color = 'hotpink')
plt.plot(low, color = 'skyblue')
plt.show()
```

```python
#실습 2번
import csv
import matplotlib.pyplot as plt

f = open('seoulTemp.csv')
#seoulTemp.csv은 1904년 1월 1일부터 2020년 12월 31일까지 서울 지역의 기온 데이터를 포함한 파일입니다.
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row [-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))

plt.hist(jan, bins=100, color='b')
plt.hist(aug, bins=100, color='r')
plt.show()
```
